#20. Grafique la curva ROC y PR del modelo

```python
from sklearn import metrics
import matplotlib.pyplot as plt
y_pred_proba = lg_.predict_proba(test_escalados_lg)[::,1]
fpr, tpr, _ = metrics.roc_curve(test_class,  y_pred_proba)
plt.plot(fpr,tpr)
plt.ylabel('Tasa de verdaderos positivos')
plt.xlabel('Tasa de falsos positivos')
plt.show()
```

```python
y_pred_proba = lg_.predict_proba(test_escalados_lg)[::,1]
fpr, tpr, _ = metrics.roc_curve(test_class,  y_pred_proba)
auc = metrics.roc_auc_score(test_class, y_pred_proba)

plt.plot(fpr,tpr,label="AUC="+str(auc))
plt.ylabel('Tasa de verdaderos positivos')
plt.xlabel('Tasa de falsos positivos')
plt.legend(loc=1)
plt.show()
```
